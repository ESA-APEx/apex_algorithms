```python
# import necessary packages
import openeo

# connect with the backend
eoconn = openeo.connect(
       "openeofed.dataspace.copernicus.eu"
).authenticate_oidc()
```

```python
# define input parameters
start_date = '2025-06-01'
end_date = '2025-07-01'

# make sure to pass cooordinate in meter and not in degree
aoi = {'east': 4880000, 'south': 2896000, 'west': 4876000, 'north': 2900000, 'crs': 3035}
process_graph = "https://raw.githubusercontent.com/ESA-APEx/apex_algorithms/obsgession_lai/algorithm_catalog/obsgession/udp_obsgession_w23_lai/openeo_udp/udp_obsgession_w23_lai.json"
```

```python
# get cube from udp
cube = eoconn.datacube_from_process(
    process_id="udp_obsgession_w23_lai",
    namespace=process_graph,
    start_date=start_date,
    end_date=end_date,
    spatial_extent=aoi,
    )
```

```python
# set the metadata for the output file
from eo_processing.utils.metadata import get_base_metadata

file_meta  = get_base_metadata(project='OBSGESSION', connection=eoconn)
file_meta.update(
    description=f'Generation of EO4Diversity conform high-resolution temperate forests optimized LAI products based on Sentinel-2 following the OBSGESSION W2.3 benchmarking.',
    tiling_grid='LAEA',
    time_start=start_date,
    time_end=end_date
)
bands_meta = {
    "LAI": {
        "description": "LAI",
        "unit": "m2*m-2",
        "valid_range": '[0, 240]',
        "scale": 1./32.,
        "offset": 0,
        "nodata_value": 255},
}
```

```python
job = cube.execute_batch(
    "./EO4Diversity_LAI_result.tiff",
    title=f'LAI calculation for {start_date} to {end_date}',
    out_format="GTiff",
    file_metadata=file_meta,
    bands_metadata=bands_meta,
    )
job
```
