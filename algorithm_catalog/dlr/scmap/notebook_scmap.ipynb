```python
import openeo

backend_url = "https://openeofed.dataspace.copernicus.eu"

connection = openeo.connect(backend_url)
connection.authenticate_oidc()
```

```python
parameters = {
    "bbox": {
        "west": 11.1,
        "south": 48.1,
        "east": 11.2,
        "north": 48.2,
        "crs": "EPSG:4326"
    },
    "temporal_extent": ["2025-04-01", "2025-05-07"],
    "max_cloud_cover": 70,
    # Optional parameters
    "nmad_sigma": 3.0,
    "max_sun_zenith_angle": 70.0,
    "compute_mref": True,
    "compute_mask": True
}
```

```python
process_graph_url = "https://raw.githubusercontent.com/ESA-APEx/apex_algorithms/refs/heads/main/algorithm_catalog/dlr/scmap/openeo_udp/scmap_composite.json"

taskmap = connection.datacube_from_process("scmap_composite", namespace=process_graph_url, **parameters)

# cube = connection.datacube_from_process(
#     process_id="scmap",
#     process_graph=process_graph_url,
#     parameters=parameters,
#     namespace="user"
# )

job = taskmap.create_job(title="scmap default Test Case")

job.start_and_wait()

results = job.get_results()
results.download_files("output")
```
