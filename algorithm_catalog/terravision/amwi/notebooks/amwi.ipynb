```python
import openeo
connection = openeo.connect("openeofed.dataspace.copernicus.eu").authenticate_oidc()
```

```python
shape = {"west": 5.05, "south": 51.21, "east": 5.1, "north": 51.23}
start_date, end_date = ["2025-05-01", "2025-05-31"]
```

```python
public_url = "https://raw.githubusercontent.com/ITA-TECNOLOGIA/Ard-Terravision/b848749026d3397cd89969e549712e21e992d4a7/src/main/python/utils/openeo_udp/amwi.json"
# Create a processing graph from the EVI process using an active openEO connection
amwi = connection.datacube_from_process(
    "AMWI", 
    namespace=public_url,
    temporal_extent=[start_date, end_date],
    spatial_extent=shape,
)
```

```python
job = amwi.create_job(
        title=f'AMWI calculation for {start_date} to {end_date}',
        out_format="NetCDF")
job.start_and_wait()
```

```python
results = job.get_results()
results.download_files()
```
